```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall = np.array([390,555,464,587,559])
years = np.array([2019,2020,2021,2022,2023])
```

```python
#line plot 
plt.plot(rainfall)
```

```python
plt.plot(years,rainfall)
```

```python
#configuring graph
plt.plot(years,rainfall)
plt.title("rainfall per years")
plt.xlabel("years")
plt.xlabel("rainfall (cm)")
plt.hlines(y=500,xmin=2019,xmax=2023,color='red',linestyles='--')
plt.show()
```

```python
plt.figure(figsize=(5,3), facecolor='lightblue')
x = np.array(['rohan','rahul','raj','ramesh'])
y = np.array([60,70,80,90])
bars = plt.bar(x,y,
        alpha=.5, edgecolor='black', linewidth = 1, hatch ='/')
bars[2].set_hatch('0')
bars[2].set_color('red')
plt.title("data")
plt.ylim(0,120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,
             bar.get_height()+10,
             f"${bar.get_height()}k",
             fontsize=10,color='red',
             ha='center')
plt.show()
```

```python
plt.figure(figsize=(5,5),facecolor='white')
plt.pie(
    y,
    labels = x,
    explode = [0,0,.06,0],
    autopct='%.1f%%',
    pctdistance=.75,
    counterclock=True,
    startangle=90,
    wedgeprops={'width':.5, 'edgecolor':'white'},
)
plt.show()
```
