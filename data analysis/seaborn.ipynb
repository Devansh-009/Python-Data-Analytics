```python
import seaborn as sns
import numpy as np
```

```python
x = np.random
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',hue='smoker',
            palette=['yellow','green'],edgecolor='black',aspect=2)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    hue="size", palette="ch:r=-.5,l=.75"
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    size="total_bill",aspect=3,
    sizes=(20,200),alpha=.5,edgecolor="blue",
)
```

```python
dowjones = sns.load_dataset("dowjones")
dowjones
```

```python
sns.relplot(data=dowjones, x="Date", y="Price", kind="line",aspect=3)
```

```python
fmri = sns.load_dataset("fmri")
fmri
```
