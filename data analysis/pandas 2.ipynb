```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
path = 'data/Canada.xlsx'
df = pd.read_excel(path, skiprows=20,sheet_name=1, skipfooter=2)
df
```

```python
pd.set_option ('display.max_columns',None) #optional
```

```python
#adding a column
years = list(range(1980, 2014))
print(years)
```

```python
df[years].sum(axis=1)
```

```python
df['total'] = df[years].sum(axis=1) 
df
```

```python
print(df.columns.tolist())
```

```python
# type, coverage,area,reg,dev
df.drop(columns=['Type','Coverage','AREA','REG','DEV'],inplace=True)
df
```

```python
df.rename(mapper={
    'OdName' : 'Country',
    'AreaName' : 'Continent',
    'RegName' : 'Region',
    'DevName' : 'Status'
},axis=1, inplace=True)
df
```

```python
#all countries with 'A'
df[df.Country.str.startswith("A")]
```

```python
#all countries ending with i using querry( )
df.query("Country.str.endswith('i')")
```

```python
# countries with total immigration less than 10000
df.query("total < 10000")
```

```python
df.set_index('Country', inplace=True)
```

```python
df.loc[['India','Japan','China'],years].T.plot(figsize=(15,5),kind='area',stacked=False,alpha=.3)
```

```python
plt.figure(facecolor='white')
colors = ['#0f8','#f00','#0f5','#0f6','#0f7','#0f4']
df.groupby('continent')['total'].sum().plot(kind='pie',autopct='%.1f%%',colors=colors,figsize=(8,8),wedgeprops={'width':.6,'edgecolor':'black',
                                                                                                  })
plt.title(f"Half of the immigrants to canada are from Asia",fontsize=20)
```

```python
sns.displot(data=df.query('total>25000'),x='total',kde=True,col='Status')
```
